```python
def remove_outlier(col):
    sorted(col)
    IQRDATA=col.quantile([.25, .5, .75]) 
    IQRDATA
    Q1=IQRDATA[0.25]
    Q2=IQRDATA[0.50]
    Q3=IQRDATA[0.75]
    IQR=Q3-Q1
    IQR
    Lower_bound=Q1-IQR*1.5
    Upeer_bound= Q3+IQR*1.5
    return Lower_bound, Upeer_bound
```

```python
Lower_bound,Upeer_bound=remove_outlier(df['lead_time'])
```

```python
Lower_bound
```

```python
Upeer_bound
```

```python
Low_Lead,Up_Lead=remove_outlier(df['lead_time'])
df['lead_time']=np.where(df['lead_time']>Up_Lead,Up_Lead,df['lead_time'])
df['lead_time']=np.where(df['lead_time']<Low_Lead,Low_Lead,df['lead_time'])
```
